```python
from recomendations.vectorizing.subscriptions import subscriptions_vector,load_default_subscription_vectorizers
```

```python
lda_vectorizer, tfidf_vectorizer, dictionary = load_default_subscription_vectorizers()
```

```python
from tqdm import tqdm
```

```python
event_ldas = dict()

for event_id, texts in tqdm(events_texts.items()):
    event_lda = subscriptions_vector([t for t in texts if t != ''], lda_vectorizer, tfidf_vectorizer, dictionary)
    event_ldas[event_id] = event_lda
```

```python
with open('resources/events/event_ldas.pkl', mode='wb') as f:
    pickle.dump(event_ldas, f)
```
