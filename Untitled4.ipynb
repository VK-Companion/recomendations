```python
import vk_api

vk_session = vk_api.VkApi(token='TOKEN_VK_TOKEN')
```

```python
vk = vk_session.get_api()
```

```python
calalog_info = vk.groups.getCatalogInfo(subcategories=True)['categories']
```

```python
calalog_info
```

```python
group_ids = []

for it in calalog_info:
    if 'subcategories' in it:
        for it_2 in it['subcategories']:
            groups = vk.groups.getCatalog(category_id=it['id'], subcategory_id=it_2['id'])['items']
    else:
        groups = vk.groups.getCatalog(category_id=it['id'])['items']
        
    for group in groups:
        group_ids.append(group['id'])
```

```python
len(group_ids)
```

```python
from tqdm import tqdm
```

```python
groups = []

for group_id in tqdm(group_ids):
    groups.append(vk.groups.getById(group_id=group_id, fields='description'))
```

```python
groups_descriptions = [group[0]['name'] + ' ' + group[0]['description'] for group in groups]
```

```python
import pymorphy2
import re
from transliterate import translit

morph = pymorphy2.analyzer.MorphAnalyzer()
alphabet = 'qwertyuiopasdfghjklzxcvbnmйцукенгшщзхъфывапролджэёячсмитьбю'


def normalize_text(s):
    s = s.lower()
    s = re.sub('https?://(-\\.)?([^\\s/?\\.#-]+\\.?)+(/[^\\s\\.]*)?', '', s)
    s = re.sub('\\b[A-Za-z0-9\\._%+-]+@[A-Za-z0-9\\.-]+\\.[A-Za-z]{2,16}\\b', '', s)
    s = ''.join([c if c in alphabet else ' ' for c in s])
    s = translit(s, 'ru')
    s = ' '.join([morph.parse(word)[0].normal_form for word in s.split(' ') if word != ''])
    return s
```

```python
with open('resources/subscriptions_vectorizers/group_descriptions.txt', mode='w') as f:
    for sent in groups_descriptions:
        f.write(sent)
        f.write('\n')
```

```python
walls = []

for group_id in tqdm(group_ids):
    walls.append(vk.wall.get(owner_id=-int(group_id), count=100, filter='owner'))
```

```python
walls[0]
```

```python
with open('resources/subscriptions_vectorizers/popular_groups.txt', mode='w') as f:
    for i in group_ids:
        f.write(str(i))
        f.write('\n')
```

```python
wall_docs = []

for wall in walls:
    for post in wall['items']:
        wall_docs.append(post.get('text', ''))
        wall_docs.append(post.get('copy_history', [{}])[0].get('text', ''))
```

```python
wall_docs
```

```python
with open('resources/subscriptions_vectorizers/popular_groups.txt', mode='w') as f:
    for doc in wall_docs:
        if doc != '':
            f.write(doc)
            f.write('\n')
```
