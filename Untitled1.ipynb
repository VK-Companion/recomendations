```python
with open('facs.txt', mode='w') as f:
    for fac in facs:
        f.write(fac)
        f.write('\n')
```

```python
import pymorphy2

morph = pymorphy2.analyzer.MorphAnalyzer()

stopwords = {'факультет', 'институт', 'бывш', 'кафедра', 'школа', 'высокий', 'имя', 'при'}


def transform_fac_name(s):
    alphabet = 'йцукенгшщзхъфывапролджэёячсмитьбю'
    s = s.lower()
    s = ''.join([c if c in alphabet else ' ' for c in s])
    s = ' '.join(word for word in s.split(' ') if len(word) >= 3)
    s = ' '.join(morph.parse(word)[0].normal_form for word in s.split(' ') if word != '')
    s = ' '.join(word for word in s.split(' ') if word not in stopwords)
    return s


transformed_facs = [transform_fac_name(fac) for fac in facs]
transformed_facs
```

```python
from collections import Counter

counter = Counter()
for fac in transformed_facs:
    for word in fac.split(' '):
        counter[word] += 1
        
counter.most_common()
```

```python
len(counter)
```
