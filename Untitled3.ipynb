```python
import re
```

```python
s = ''
```

```python
import pymorphy2

morph = pymorphy2.analyzer.MorphAnalyzer()
alphabet = 'qwertyuiopasdfghjklzxcvbnmйцукенгшщзхъфывапролджэёячсмитьбю'


def normalize_text(s):
    s = s.lower()
    s = re.sub('https?://(-\\.)?([^\\s/?\\.#-]+\\.?)+(/[^\\s\\.]*)?', '', s)
    s = re.sub('\\b[A-Za-z0-9\\._%+-]+@[A-Za-z0-9\\.-]+\\.[A-Za-z]{2,16}\\b', '', s)
    s = ''.join([c if c in alphabet else ' ' for c in s])
    s = translit(s, 'ru')
    s = ' '.join([morph.parse(word)[0].normal_form for word in s.split(' ') if word != ''])
    return s
```

```python
sub_descs = [normalize_text('{}. {}'.format(sub.get('name', ''), sub.get('description', ''))) for sub in subs]
```
